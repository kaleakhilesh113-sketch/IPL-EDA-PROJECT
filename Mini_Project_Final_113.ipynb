DLS method for rainfall

```python
import matplotlib.pyplot as plt

dl_counts = df['dl_applied'].value_counts()

plt.pie(dl_counts,
        labels=["Not Applied (0)", "Applied (1)"],
        autopct='%1.1f%%',
        colors=["#66b3ff", "#ff9999"],
        startangle=90)

plt.title("DLS Method Application in IPL Matches")
plt.show()
```

Top 5 venue According Team performance on it venue

```python
top_venues = df['venue'].value_counts().head(5).index
venue_team_wins = df[df['venue'].isin(top_venues)].groupby(['venue', 'winner']).size().unstack().fillna(0)

plt.figure(figsize=(12, 6))
sns.heatmap(venue_team_wins, annot=True, fmt=".0f", cmap='YlGnBu')
plt.title("Team Wins at Top IPL Venues")
plt.xlabel("Team")
plt.ylabel("Venue")
plt.show()
```
